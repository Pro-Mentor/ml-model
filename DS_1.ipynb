```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# to preprocess text data
import string
import nltk
# from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('wordnet')

from textblob import Word
from wordcloud import WordCloud
```

```python
path = "/content/drive/MyDrive/FYP/datasets/data-engineer-jobs.csv"
df = pd.read_csv(path)
df.head(5)
```

```python
df_tmp = pd.DataFrame(columns=['title', 'description'])

df_tmp['title'] = df['Job Title']
df_tmp['description'] = df['Job Description']
df_tmp.head().T
```

```python
# Check the values of different columns

# Set the display option to show all rows
pd.set_option('display.max_rows', None)

df_tmp.title.value_counts()
```

```python
df_tmp.replace({'Big Data Engineer': 'Data Engineer', 'Senior Data Engineer': 'Data Engineer'}, inplace=True)
df_tmp.head().T

df_ds = df_tmp[df_tmp['title'] == 'Data Engineer'][['title', 'description']]
len(df_ds)
```
